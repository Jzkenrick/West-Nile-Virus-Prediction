```python
# define function to return the gs cv scores for the best parameter
def get_mean_gs_cv_score(gs, splits = 5):
    best_param_index = np.where(gs.cv_results_['rank_test_score']==1)[0][0]
    scores = []
    for split in np.arange(splits):
        split_score = gs.cv_results_[f'split{split}_test_score'][best_param_index]
        scores.append(split_score)
    return scores
```

#### GridSearch iteration 1

```python
# GS iteration 1
xgb = XGBClassifier(learning_rate=0.1,
                    n_estimators=500,
                    min_child_weight=1,
                    subsample=0.8,
                    colsample_bytree=0.8,
                    scale_pos_weight=19, # ratio of neg instances/pos instances
                    seed=27)

pl = Pipeline([ 
    ('classifier', xgb)
])

params_1 = {
    'classifier__max_depth': [5, 10],
    'classifier__gamma': [0.5, 1, 3],
    'classifier__reg_lambda': [2, 5],
    'classifier__reg_alpha': [1.5, 2, 3]
}

gs_1 = GridSearchCV(pl, param_grid = params_1, scoring='roc_auc', n_jobs=-2, verbose=True, cv=5)
gs_1.fit(xtrain, ytrain)

print(f'mean CV score: {get_mean_gs_cv_score(gs_1)}')
print(f'best gridsearch parameters: {gs_1.best_params_}')

%timeit
```

```python
def fit_xgb(model, train_df, predictor_var, target_var):
    # get param of instantiated xgb model
    xgb_param = model.get_xgb_params()
    # put data into a DMatrix
    xgb_train = xgb.DMatrix(train_df[predictor_var].values, label=train_df[target_var].values)
    # carry out xgb cv (on each stage of tree growth)
    cv_result = xgb.cv(xgb_param, xgb_train, 
                      num_boost_round=model.get_params()['n_estimators'], 
                      nfold=5,
                      metrics='auc', 
                      early_stopping_rounds=50)
    
    # update model param with cv result
    model.set_params(n_estimators=cv_result.shape[0])
    # fit updated model to training data
    model.fit(train_df[predictor_var], train_df[target_var],eval_metric='auc')
        
    # run predictions
    ypred = model.predict(train_df[predictor_var])
    pred_proba = model.predict_proba(train_df[predictor_var])[:,1]
        
    # print metrics
    print (f'AUC Score (train): {roc_auc_score(train_df[target_var], ypred)}')
    
    # plot feature importances
    feat_imp = pd.DataFrame(model.feature_importances_, index = xtrain_balanced.columns).sort_values(0,ascending=False)
    plt.figure(figsize=(10,7))
    sns.barplot(x=feat_imp[0], y=feat_imp.index, color='skyblue')
    plt.title('Feature importance scores', fontsize=14)
    plt.xlabel('')
    plt.yticks(fontsize=12)
    plt.tight_layout()
```

## Automated feature engineering

### Convert data to int/float32

```python
def convert_columns(df):  
    for col in df.columns:    
        col_type = df[col].dtype
        if (col_type == 'object'):
                df[col] = df[col].astype('category')    
        if (col_type == 'int64'):
                df[col] = df[col].astype('int32')   
        if (col_type == 'float64'):
                df[col] = df[col].astype('float32')  
    return df

for col in train:
    convert_columns(train)
    
train.info()
```
